```python
from matplotlib import rc
from matplotlib.pyplot import plot, xlabel, ylabel, legend, rcParams
from numpy import linspace
from sympy import *
# sympy pretty printing
init_printing()
rcParams['figure.figsize'] = (15, 15)

font = {'family' : 'Liberation Sans',
        'weight' : 'bold',
        'size'   : 32}
rc('font', **font)

x, y, z, s, t, u, v, theta, phi, rho = symbols(r'x y z s t u v \theta \phi \rho', real=True)
m, V = symbols(r'm V', real=True)
r = Symbol('r', nonzero=True, positive=True)

def plot_eqn_contours(eqn, levels=None, x_limits=(-5, 5), y_limits=(-5, 5), resolution=100, show=True):
    x_vals = linspace(x_limits[0], x_limits[1], resolution * (x_limits[1] - x_limits[0]))
    y_vals = linspace(y_limits[0], y_limits[1], resolution * (y_limits[1] - y_limits[0]))
    xy_vals = np.meshgrid(x_vals, y_vals)
    eqn_lambda = lambdify((x, y), eqn, modules='numpy')
    cs = contour(xy_vals[0], xy_vals[1], eqn_lambda(xy_vals[0], xy_vals[1]), levels=levels)
    cs.clabel();
    return cs
```

# Critical Point Classification from Contour Plots of Level Curves

Contour plots of level curves provide a nice way of understanding critical points and their classification.

* Points which are completely surrounded by a contour are going to have a local max or min.

* Points which are at the intersection of two contours are generally saddles; though the regions around it must alternate from increasing to decreasing in height.

Identify all of the critical points by inspection in the plots below.

```python
plot_eqn_contours(x ** 2 + y ** 2 , x_limits = (-3, 3), y_limits = (-5, 5));
```

```python
plot_eqn_contours(sin(x) + cos(y), levels = np.arange(-2.0, 2.0, 0.5), x_limits=(-np.pi, np.pi), y_limits=(-np.pi, np.pi));
```

```python
plot_eqn_contours(sin(x) * cos(y), levels = np.arange(-1.0, 1.0, 0.25), x_limits=(-np.pi, np.pi), y_limits=(-np.pi, np.pi));
```
